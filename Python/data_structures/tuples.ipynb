# Tuples

### Creating Tuples

```python
empty_tuple = ()
print("empty tuple", empty_tuple)

one_element_tuple = (42,) # trailing comma is required for single element tuple
print("one element tuple", one_element_tuple)

#heterogeneous tuple
heterogeneous_tuple = (1, "hello", 3.14)
print("heterogeneous tuple", heterogeneous_tuple)
print("length of heterogeneous tuple", len(heterogeneous_tuple))

# tuple without parentheses
tuple_without_parentheses = 1, "world", 2.718
print("tuple without parentheses", tuple_without_parentheses)
```

### Accessing Tuple elements

```python
rgb_colors = (255, 0, 0) # red color in RGB
print("RGB colors tuple", rgb_colors)
print("red:", rgb_colors[0])
print("green:", rgb_colors[1])
print("blue:", rgb_colors[2])

red, green, blue = rgb_colors
print("unpacked RGB colors:", red, green, blue)
```

### Reverse tuple  

```python
steps = ("start","load","process","end")
print("steps tuple", steps)

# reversing a tuple
reversed_steps = tuple(reversed(steps))
print("reversed steps tuple", reversed_steps)

# reversing a tuple using slicing
reversed_steps_slicing = steps[::-1]
print("reversed steps tuple using slicing", reversed_steps_slicing)
```

### Sorting tuple

```python
random_numbers = (5, 2, 9, 1, 15, 6)
print("random numbers tuple", random_numbers)

sorted_numbers = tuple(sorted(random_numbers))
print("sorted numbers tuple", sorted_numbers)

# sorting a tuple with key function
records = [("Alice", 30), ("Bob", 25), ("Charlie", 35)]
print("records list", records)
sorted_records = tuple(sorted(records, key = lambda x:x[1]))
print("sorted records tuple", sorted_records)
```

```python
sample = ["(1,2)", "(2,4)", "(5,7)", "(7,2)", "(9,5)"]

def validate_binary_tree(binary_tree):
    parent_child_dict = {
        "parents": {},
        "childrenCount": {}
    }

    for item in binary_tree:
        child, parent = tuple(map(int, item.strip("()").split(",") ))

        # rule 1: child should have only one parent
        if child in parent_child_dict["parents"]:
            return False
        parent_child_dict["parents"][child] = parent

        # rule 2: parent can have at most two children
        if parent in parent_child_dict["childrenCount"]:
            parent_child_dict["childrenCount"][parent] += 1
            if parent_child_dict["childrenCount"][parent] > 2:
                return False
        else:
            parent_child_dict["childrenCount"][parent] = 1

    return True

validate_binary_tree(sample)
```

```python
# formula 2n! / (n+1)!n! for number of unique binary trees with n nodes
def factorial(n):
    if n == 0 or n == 1:
        return 1
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

def catalan_number(n):
    if n == 0:
        return 1
    return factorial(2 * n) // (factorial(n + 1) * factorial(n))
print("number of unique binary trees with 3 nodes:", catalan_number(8))
```

```python
def FindIntersection(strArr):

  # code goes here
  first_element,second_element = strArr
  first_element_list = list(first_element.split(','))
  second_element_list = list(second_element.split(','))

  output=''
  for i in second_element_list:
    if(i in first_element_list):
      if(len(output) == 0):
        output = i
      else:
        output += f",{i}"
  return output

# keep this function call here 
print(FindIntersection(["1, 3, 4, 7, 13", "1, 2, 4, 13, 15"]))
```
