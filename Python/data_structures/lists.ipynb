# Lists

### Creating Lists:

```python
# Empty list
empty_list = []

# Homogeneous list
numbers = [1, 2, 3, 4, 5]
# Heterogeneous list
mixed_list = [1, "Hello", 3.14, True]
# Nested list
nested_list = [1, [2, 3], [4, 5, 6]]

# Accessing elements
first_number = numbers[0]  # Accessing the first element

print("Empty List:", empty_list)
print("Homogeneous List:", numbers)
print("Heterogeneous List:", mixed_list)
print("Nested List:", nested_list)
print("First number in the homogeneous list:", first_number)
print("Length of the homogeneous list:", len(numbers))
print("last element of the heterogeneous list:", mixed_list[-1])
print("negative indexing in the nested list:", mixed_list[-2])
```

### List Mutation

```python
numbers_list = [1, 2, 3, 4, 5]
# Modifying lists
numbers_list.append(6)  # Adding an element to the end of the list
print("Modified Homogeneous List:", numbers_list)

numbers_list.insert(0, 0)  # Inserting an element at the beginning of the list
print("Inserting an element at the beginning of the list:", numbers_list)

numbers_list[2] = 10  # Modifying an element at a specific index
print("Modifying an element at a specific index:", numbers_list)

# Removing elements
numbers_list.remove(10)  # Removing a specific element by value
print("Removing a specific element by value:", numbers_list)

numbers_list.pop(0)  # Removing an element by index (removes the first element)
print("After Removing Elements:", numbers_list)

# remove all the elements from the list
numbers_list.clear()
print("After Removing All Elements:", numbers_list)
```

### Concatation

```python
# current queue
print_queue = ["Invoice.pdf", "Report.docx", "Presentation.pptx"]
print("Print Queue:", print_queue)

# Adding a new document to the print queue
new_queue = ["Summary.xlsx"]

print_queue.extend(new_queue)  # Adding the new document to the print queue
print("Updated Print Queue:", print_queue)

# inseting a document at the beginning of the print queue
print_queue.insert(0, "Agenda.pdf")
print("After Inserting a Document at the Beginning of the Print Queue:", print_queue)

# concatenating two lists
additional_queue = ["Notes.txt", "Budget.xlsx"]
full_queue = print_queue + additional_queue  # Concatenating the two lists
print("Full Print Queue:", full_queue)
```

### Iterating through a list

```python
# iterating through a list
list_of_tasks = ["Task 1", "Task 2", "Task 3"]

for task in list_of_tasks:
    print("Current Task:", task)

# iterating through a list with index
for index, task in enumerate(list_of_tasks):
    print(f"{index + 1}: {task}")
```

### Counting in lists

```python
# counting in lists
fruits = ["apple", "banana", "orange", "apple", "grape", "banana"]
apple_count = fruits.count("apple")  # Counting the occurrences of "apple"
print("Number of times 'apple' appears in the list:", apple_count)

# index of an element in the list
banana_index = fruits.index("banana")  # Finding the index of the first occurrence of "banana"
print("Index of the first occurrence of 'banana':", banana_index)

# Check if an element exists in the list
if "grape" in fruits:
    print("Grape is found at index:", fruits.index("grape"))
else:
    print("Grape is not in the list of fruits.")

# Check if an element does not exist in the list
if "kiwi" not in fruits:
    print("Kiwi is not in the list of fruits.")
```

### Sorting with List In-Place Sort

```python
# sorting a list with the sort() method
numbers_to_sort = [5, 2, 9, 1, 5, 6]
print("Original Numbers:", numbers_to_sort)
numbers_to_sort.sort()  # Sorting the list in ascending order
print("Sorted Numbers:", numbers_to_sort)

# sorting a string list
names = ["Charlie", "Alice", "Bob"]
print("Original Names:", names)
names.sort()  # Sorting the list in alphabetical order
print("Sorted Names:", names)

# sorting a list in reverse order
numbers_to_sort.sort(reverse=True)  # Sorting the list in descending order
print("Sorted Numbers in Reverse Order:", numbers_to_sort)

# sorting a list with the sorted() function
unsorted_numbers = [5, 2, 9, 1, 5, 6]
sorted_numbers = sorted(unsorted_numbers)  # Sorting the list and creating a new sorted list
print("Original Unsorted Numbers:", unsorted_numbers)
print("New Sorted Numbers:", sorted_numbers)

# Flipping a list with reversed() function
original_list = [1, 2, 3, 4, 5]
flipped_list = list(reversed(original_list))  # Flipping the list using reversed() function
print("Original List:", original_list)
print("Flipped List:", flipped_list)
```

### Slicing a List

```python
# slicing a list
list_to_slice = [1, 2, 3, 4, 5]
sliced_list = list_to_slice[1:4]  # Slicing the list from index 1 to index 3 (4 is exclusive)
print("Original List:", list_to_slice)
print("Sliced List (index 1 to 3):", sliced_list)

# slicing a list with negative indices
sliced_list_negative = list_to_slice[-4:-1]  # Slicing the list
print("Sliced List with Negative Indices (index -4 to -2):", sliced_list_negative)

# slicing a list with step
sliced_list_step = list_to_slice[0:5:2]  # Slicing
print("Sliced List with Step (index 0 to 4 with step 2):", sliced_list_step)
```

### Listing Alias

```python
# listing with alias
original_list = [1, 2, 3, 4, 5]
alias_list = original_list  # Creating an alias for the original list
alias_list.append(6)  # Modifying the alias list
print("Original List after modifying alias:", original_list)
print("Alias List:", alias_list)

# shallow copy of a list
original_list = [1, 2, 3, 4, 5]
shallow_copied_list = original_list.copy()  # Creating a shallow copy of the original list
shallow_copied_list.append(6)  # Modifying the shallow copied list
print("Original List after modifying shallow copy:", original_list)
print("Shallow Copied List:", shallow_copied_list)
```

### Shallow copy vs Deep copy

```python
# shallow copy with slicing
settings = [["volume", 70], ["brightness", 50]]
shallow_copied_settings = settings[:]  # Creating a shallow copy using slicing
shallow_copied_settings[0][1] = 80  # Modifying the nested list in the shallow copy
print("Original Settings after modifying shallow copy:", settings)
print("Shallow Copied Settings:", shallow_copied_settings)

# deep copy of a list
import copy
original_settings = [["volume", 70], ["brightness", 50]]
deep_copied_settings = copy.deepcopy(original_settings)  # Creating a deep copy of the original list
deep_copied_settings[0][1] = 80  # Modifying the nested list in the deep copy
print("Original Settings after modifying deep copy:", original_settings)
print("Deep Copied Settings:", deep_copied_settings)
```

### List Unpacking

```python
# list unpacking
numbers = [1, 2, 3]
a, b, c = numbers  # Unpacking the list into individual variables
print("Unpacked Variables:", a, b, c)

# list unpacking with nested lists
nested_numbers = [1, [2, 3], 4]
a, (b, c), d = nested_numbers  # Unpacking the nested list
print("Unpacked Variables from Nested List:", a, b, c, d)

# list unpacking with extended unpacking
numbers = [1, 2, 3, 4, 5]
a, b, *rest = numbers  # Unpacking with extended unpacking
print("Unpacked Variables with Extended Unpacking:", a, b, rest)

# Transforming a list with list comprehension
squared_numbers = [x**2 for x in range(1, 6)]
print("Squared Numbers:", squared_numbers)
```

```python
numbers = [1, 2, 3, 4, 5]

# Filtering a list with list comprehension
odd_numbers = [n for n in numbers if n % 2 != 0]

print("Odd Numbers:", odd_numbers)
```

### Joining Lists

```python
# joining two lists with list comprehension
folders = ["Documents", "Pictures", "Music"]
print("Folders:", folders)

folder_paths = [f"/home/muthu/{folder}" for folder in folders]
print("Folder Paths:", folder_paths)

# by using join() method
folder_paths_joined = ", ".join(folder_paths)
print("Folder Paths Joined:", folder_paths_joined)
```
