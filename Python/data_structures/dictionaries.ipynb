# Dictionaries

### Creating dictionaries

```python
empty_dict = {}
print("empty_dict:", empty_dict)

setting_dict = {
    "resolution": "1920x1080",
    "fullscreen": True,
    "volume": 75,
}

print("setting_dict:", setting_dict)
print("Keys:", setting_dict.keys())
print("Values:", setting_dict.values())
print("Items:", setting_dict.items())
print("get length:", len(setting_dict))
```

### Initializing key with fromkeys() method

```python
permissions = ['read', 'write', 'execute', 'delete']
permissions_dict = dict.fromkeys(permissions, False)
print("permissions_dict:", permissions_dict)
print("access 'write':", permissions_dict['write'])
```

### Key error

```python
setting_dict = {
    "resolution": "1920x1080",
    "fullscreen": True,
    "volume": 75,
}

try:
    print("setting_dict:", setting_dict['brightness'])
except KeyError:
    print("Brightness key not found:")
```

### Accession elements with get()

```python
config = {
    "resolution": "1920x1080",
    "fullscreen": True,
}

resolution = config.get("resolution", "[missing]: set default value")
print("resolution:", resolution)

brightness = config.get("brightness", "[missing]: set default value")
print("brightness:", brightness)

theme = config.get("theme")
print("theme:", theme)
```

### Apply default value with setDefault()   

```python
user = {
    "username": "john_doe",
}

user_role = user.setdefault("role", "viewer")
print("user_role:", user_role)

user_name = user.setdefault("username", "default_user")
print("dict:", user)
```

### Updating dictionary with update()

```python
global_config = {
    "resolution": "1920x1080",
    "fullscreen": True,
}

device_config = {
    "fullscreen": False,
    "volume": 75,
}

global_config.update(device_config)
print("global_config:", global_config)

# merge update method
mobile_config = {
    "resolution": "1280x720",
    "fullscreen": True,
    "volume": 50,
}

global_config |= mobile_config
print("global_config after merge update:", global_config)

new_config = global_config | mobile_config
print("new_config after merge:", new_config)

```

### Remove an item

```python
# using del to remove a key-value pair from the dictionary
global_config = {
    "resolution": "1920x1080",
    "fullscreen": True,
    "volume": 75,
}

print("Before deletion:", global_config)

del global_config["fullscreen"]

print("After deletion:", global_config)

# using pop to remove a key-value pair and return the value
global_config = {
    "resolution": "1920x1080",
    "fullscreen": True,
    "volume": 75,
}
print("Before pop:", global_config)
fullscreen_value = global_config.pop("fullscreen", None) # provide default value to avoid KeyError if key is not found
print("Popped value:", fullscreen_value)
print("After pop:", global_config)

# using popitem to remove and return the last inserted key-value pair
global_config = {
    "resolution": "1920x1080",
    "fullscreen": True,
    "volume": 75,
}
print("Before popitem:", global_config)
last_item = global_config.popitem()
print("Popped item:", last_item)
print("After popitem:", global_config)

# using clear to remove all key-value pairs from the dictionary
global_config = {
    "resolution": "1920x1080",
    "fullscreen": True,
    "volume": 75,
}
print("Before clear:", global_config)
global_config.clear()
print("After clear:", global_config)
```

### Sorting

```python
prices = {
    "keyboard": 49.99,
    "mouse": 19.99,
    "chair": 89.99,
    "monitor": 199.99,
}

print(prices.items())

# sorting by keys
sorted_prices = dict(sorted(prices.items()))
print("Sorted prices:", sorted_prices)

# sorting by values
sorted_prices_by_value = dict(sorted(prices.items(), key=lambda item: item[1]))
print("Sorted prices by value:", sorted_prices_by_value)

# sorting with itemgetter
from operator import itemgetter
sorted_prices_by_value_itemgetter = dict(sorted(prices.items(), key=itemgetter(1)))
print("Sorted prices by value using itemgetter:", sorted_prices_by_value_itemgetter)
```
